```python
col_list_new = list(set(col_list) - set(['_c0', 'Polygon_ID', 'acq_date', 'frp']))
```

```python
from pyspark.sql.functions import col,isnan, when, count
test.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in col_list_new]
   ).show()
```

```python
import pyspark.sql.functions as f
from functools import reduce

try_null = reduce(
    lambda a,b: a.filter(test[b].isNotNull()),
    test.columns,
    test
)
```

```python
try_null.show()
```

```python
try_null.where(reduce(lambda x, y: x | y, (f.col(x).isNull() for x in test.columns))).show()
```
