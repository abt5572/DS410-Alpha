```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('wildfiredb.csv')
```

```python
data['year'] = pd.to_datetime(data['acq_date']).dt.year
data['month'] = pd.to_datetime(data['acq_date']).dt.month
```

```python
heatmap_data = data.pivot_table(index='year', columns='month', values='frp', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlOrRd', annot=True, fmt=".2f", linewidths=.5)
plt.title("Average FRP by Month and Year")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=data, x='year', y='frp')
plt.title("Distribution of FRP by Year")
plt.ylabel("FRP")
plt.xlabel("Year")
plt.show()

plt.figure(figsize=(14, 6))
sns.boxplot(data=data, x='month', y='frp')
plt.title("Distribution of FRP by Month")
plt.ylabel("FRP")
plt.xlabel("Month")
plt.show()
```
